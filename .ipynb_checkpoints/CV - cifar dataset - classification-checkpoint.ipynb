```python
data = {'x_train':[], 'y_train': []}
for i in total_batches:
    for j in i[b'data']:
        arr = j.reshape(32,32,3)
        data['x_train'].append(arr)
for i in total_batches:
    for j in i[b'labels']:
        data['y_train'].append(j)
```

```python
train_examples = data['x_train']
train_labels = data['y_train']
```

```python
test_batch = batch = unpickle('.\cifar-10-batches-py\test_batch')
```

```python
train_labels = np.array(train_labels)
```

```python
train_labels = np.expand_dims(train_labels,axis=1)
train_labels[0]
```

```python
train_labels = tf.keras.utils.to_categorical(
    train_labels, num_classes=10, dtype='float32'
)
```

```python
train_labels[0]
```

```python
all_dataset = tf.data.Dataset.from_tensor_slices((train_examples, train_labels))
validation_set = all_dataset.take(1000)
train_dataset = all_dataset.skip(1000)

BATCH_SIZE = 64
SHUFFLE_BUFFER_SIZE = 100

train_dataset = train_dataset.shuffle(SHUFFLE_BUFFER_SIZE).batch(BATCH_SIZE)
validation_set = validation_set.batch(BATCH_SIZE)
```

```python
train_dataset
```

```python
validation_set
```

```python
model = Sequential([\
                    Conv2D(filters=32, kernel_size=(3,3), activation='relu', padding = 'same', input_shape=(32,32,3)),
                    MaxPool2D(pool_size=(2,2), strides=2),\
                    Conv2D(filters=64, kernel_size=(3,3), activation='relu', padding='same'),\
                    MaxPool2D(pool_size=(2,2), strides=2),\
                    Flatten(),\
                    Dense(units=10, activation='softmax')])
```

```python
model.summary()
```

```python
model.compile(optimizer=Adam(learning_rate=0.0001), loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
model.fit(x=train_dataset,validation_data = validation_set, epochs=100, verbose=2)
```
